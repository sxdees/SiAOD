### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со
временем стандартной функции сортировки. Испытания проводить на сгенерированных
матрицах.

```python
import copy
import time
import random
import heapq

def generate_matrix(m, n, min_limit, max_limit):
    e=[]
    for b in range(m):
        row=[]
        for c in range(n):
            ch=random.randint(min_limit,max_limit)
            row.append(ch)
        e.append(row)
    return e

user_m=input('Введите число строк ')
user_n=input('Введите число столбцов ')
user_min_limit=input('Введите нижнюю границу ')
user_max_limit=input('Введите верхнюю границу ')

user_m=int(user_m)
user_n=int(user_n)
user_min_limit=int(user_min_limit)
user_max_limit=int(user_max_limit)

if user_min_limit > user_max_limit:
    print('Вы ввели не то!')
else:
    e=generate_matrix(user_m, user_n, user_min_limit, user_max_limit)
    for row in e:
        print(row)

start_time=time.time()
for row in e:
    row.sort()
    print(row)
print()    
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка выбором.
start_time = time.time()

def sortvyb(a):
    for i in range(len(a)):
        m_ind=i
        for k in range(i+1, len(a)):
            if a[k] < a[m_ind]:
                m_ind=k
        a[i], a[m_ind] = a[m_ind], a[i]

for row in e:
    sortvyb(row)
    print(row)

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка вставкой.
start_time = time.time()

def sortvstv(arr):
    for i in range(1, len(arr)):
        now=arr[i]
        b=i-1
        while b>=0 and arr[b] > now:
            arr[b+1]=arr[b]
            b=b-1
            arr[b+1]=now

for row in e:
    sortvstv(row)
    print(row)

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка обменом. || Сортировка пузырьком.
start_time = time.time()

def sortpuz(arr):
    a=len(arr)
    for i in range(a):
        for k in range(0, a-i-1):
            if arr[k] > arr[k+1]:
                arr[k], arr[k+1]=arr[k+1], arr[k]

for row in e:
    sortpuz(row)
    print(row)

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка Шелла.
start_time = time.time()

def sortshell(arr):
    a=len(arr)
    gap=a//2
    while gap>0:
        for i in range(gap, a):
            c=arr[i]
            b=i
            while b>=gap and arr[b-gap]>c:
                arr[b]=arr[b-gap]
                b=b-gap
                arr[b]=c
        gap=gap//2

for row in e:
    sortshell(row)
    print(row)

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Быстрая сортировка.
start_time = time.time()

def sortbyst(arr):
    if len(arr) <=1:
        return arr
    pivot=arr[len(arr) // 2]
    l=[x for x in arr if x < pivot]
    m=[x for x in arr if x==pivot]
    r=[x for x in arr if x > pivot]
    return sortbyst(l)+m+sortbyst(r)

for row in e:
    sortbyst(row)
    print(row)

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
    
```

```python
# Турнирная сортировка.
start_time = time.time()

def sorttour(arr):
    heap = list(arr)
    heapq.heapify(heap)
    sorted_arr = [heapq.heappop(heap) for _ in range(len(arr))]
    return sorted_arr

for row in e:
    sorttour(row)
    print(row)
    
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```
